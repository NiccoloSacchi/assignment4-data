# Text language classification with Fasttext model

```python
from cs336_data.data_processing import get_fasttext_model, ModelType

m = get_fasttext_model(ModelType.LANG_CLASSIFIER)
m.predict(["My nome è Niccolo", "Ich bin 20 yahre alt"], k=1)
```

```python
from cs336_data.data_processing import extract_text_from_warc, identify_language

texts = extract_text_from_warc("../local-shared-data/CC/example.warc.gz", n=100)

for text in texts:
    text_str = text["text"]
    lan, score = identify_language(text_str)
    print(f"{lan}|{score}: {text_str[:100]}")
    print("-------------------------------------------------------------------")
```

# Personally identifiable information (PII) masking

# Harmful content classification

```python
from cs336_data.data_processing import get_fasttext_model, ModelType

nsfw_model = get_fasttext_model(ModelType.NSFW_CLASSIFIER)
hate_model = get_fasttext_model(ModelType.HATE_SPEECH_CLASSIFIER)

texts = ["fuck you", "die bitch", "I love you", "Good job, jerk"]
nsfw_model.predict(texts), hate_model.predict(texts)
```
